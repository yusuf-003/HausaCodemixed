```python
# Example: Assuming you have uploaded a CSV file with words and their expected stems
# The CSV should have two columns: 'Word' and 'ExpectedStem'
file_path = 'words_and_stems_simple.csv'  # Update with the actual path to your file

# Load the data
try:
    data = pd.read_csv(file_path)
except FileNotFoundError:
    raise FileNotFoundError(f"File not found at {file_path}. Please ensure the file path is correct.")

# Ensure the necessary columns are in the file
if not {'word', 'expected_stem_simple'}.issubset(data.columns):
    raise ValueError("The CSV file must contain 'word' and 'ExpectedStem' columns.")

# Initialize your HausaStemmer instance
stemmer = HausaStemmer()

# Add a column for the stemmer output and correctness
data['StemmerOutput'] = data['word'].apply(lambda word: stemmer.stem(word))
data['Correct'] = data['StemmerOutput'] == data['expected_stem_simple']

# Calculate accuracy metrics
total_words = len(data)
correct_stems = data['Correct'].sum()
accuracy = (correct_stems / total_words) * 100

# Display results
print(f"Total Words Evaluated: {total_words}")
print(f"Correctly Stemmed Words: {correct_stems}")
print(f"Accuracy: {accuracy:.2f}%")

# Save the evaluation results to a new CSV file
evaluation_file_path = 'hausastemmer_evaluation_simple_results.csv'
data.to_csv(evaluation_file_path, index=False)
```
